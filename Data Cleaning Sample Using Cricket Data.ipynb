```python
#Import Pandas for data Clean
import pandas as pd
```

```python
#Import the data
cricket_data=pd.read_csv('C:\\Users\\Brian Motee\\Downloads\\Decibels\\Cricket_Data.csv')
#Display top 5 rows
cricket_data.head()
```

```python
#Renaming Multiple Columns
cricket_data=cricket_data.rename(columns={'Mat':'Matches','NO':'Not_Outs','HS':'Highest_Inns_Score','BF':'Balls_Faced','SR':'Batting_Strike_Rate'})
#Inspect if Changes were made
cricket_data.head()
```

```python
#Checking for null values
cricket_data.isnull().any()
```

```python
#Checking specific rows with null values
cricket_data[cricket_data['Balls_Faced'].isna()==1]
```

I could fill null values with the next value to it(bfill), the value before it(ffill) or with zeroes. In this case I chose to fill the null values with zeroes instead bfill or ffill as I don't assume the missing values will remain contanst untill the next observation

```python
#Filling Missing values with zeroes
cricket_data['Balls_Faced']=cricket_data['Balls_Faced'].fillna(0)
cricket_data['Batting_Strike_Rate']=cricket_data['Batting_Strike_Rate'].fillna(0)
#Inspecting for missing values
cricket_data[cricket_data['Player']=='Hon.FS Jackson (ENG)']
```

```python
#Checking for duplicates
cricket_data.duplicated()
```

```python
#Checking the duplicated rows
cricket_data[cricket_data['Player'].duplicated()==1]
```

```python
#Investigating further the duplicated rows
cricket_data[cricket_data['Player'].isin(['GA Headley (WI)','H Sutcliffe (ENG)','E Paynter (ENG)','KF Barrington (ENG)','ED Weekes (WI)','WR Hammond (ENG)'])]
```

Since the rows are exact duplicates without a logical reason behind it, I will drop the duplicates

```python
#Dropping Duplicates
cricket_data.drop_duplicates()
```

The number of rows reduced from 66 to 62 indicating the duplicates were dropped

```python
#Splitting the Span Column into Start_Year and End_Year
cricket_data['Start_Year']=cricket_data['Span'].str.split(pat='-').str[0]
cricket_data['End_Year']=cricket_data['Span'].str.split(pat='-').str[1]
#Inspecting the data for the changes
cricket_data.head()
```

I don't have the use for the Span column anymore, so I will drop it

```python
#Dropping Span Column
cricket_data=cricket_data.drop(['Span'],axis=1)
```

```python
#Inspecting for changes
cricket_data.head()
```

```python
#Splitting the player column to column name and country
cricket_data['Player'].str.split(pat='(')
```

```python
cricket_data['Country']=cricket_data['Player'].str.split(pat='(').str[1]
```

```python
cricket_data['Country']=cricket_data['Country'].str.split(pat=')').str[0]
```

```python
cricket_data['Country']
```

```python
cricket_data['Player']=cricket_data['Player'].str.split(pat='(').str[0]
```

```python
cricket_data.head()
```

```python
#inspecting the data types
cricket_data.dtypes
```

From the results, the Matches, Highest_Inns_Score, Start_Year and End_Year will need to be converted from object to Int but its not straight forward. for instance I will need to remove the (*) symbol from values in the Highest_Inns_Score  column

```python
#Removing the * from  Highest_Inns_Score column
cricket_data['Highest_Inns_Score'].str.split(pat='*').str[0]
#Assign it
cricket_data['Highest_Inns_Score']=cricket_data['Highest_Inns_Score'].str.split(pat='*').str[0]
```

```python
#Changing it to Integer
cricket_data['Highest_Inns_Score']=cricket_data['Highest_Inns_Score'].astype('int')
```

```python
#Changing Start_Year and End_Year to integer
cricket_data=cricket_data.astype({'Start_Year':'int','End_Year':'int'})
```

```python
#Changing Matches to Integer
cricket_data['Matches']=cricket_data['Matches'].str.split(pat='*').str[0]
#Converting matches
cricket_data['Matches']=cricket_data['Matches'].astype('int')
```

```python
#Calculating the career length based on End_Year and Start_Year
cricket_data['Career_Length']=cricket_data['End_Year']-cricket_data['Start_Year']
cricket_data
```

I want to perform some analysis to test if the dataset is analysis ready

```python
#What is the average career length of cricket players?
cricket_data['Career_Length'].mean()
```

```python
#What is the highest score by country
cricket_data.groupby('Country')['Highest_Inns_Score'].max().to_frame('High_Inn_Score').reset_index().sort_values('High_Inn_Score',ascending=False)
```
